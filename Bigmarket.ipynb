# Preprocessing

Categorical:

```python
df_cat=df[cat].drop('Outlet_Establishment_Year',axis=1)
```

```python
df_cat_dummies=pd.get_dummies(df_cat,dtype=int)
```

```python
df_cat_dummies.shape
```

Numerical:

```python
df['Item_Identifier'].nunique()
```

```python
df_num=df[cont].drop(['Item_Identifier','Item_Outlet_Sales'],axis=1)
```

```python
# Apply Min Max Scaler
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MM=MinMaxScaler()
```

```python
df_num=df_num.drop('Item_Type',axis=1)
```

```python
df_num_mm=MM.fit_transform(df_num)
```

```python
df_num_df=pd.DataFrame(df_num_mm,columns=df_num.columns)
```

```python
df_final=pd.concat([df_num_df,df_cat_dummies],axis=1)
```

```python
df_final
```

```python
df_final.shape
```

# 5. Decide I/p and O/p

```python
X=df_final 
Y=df['Item_Outlet_Sales']
```

# 6. Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.10,random_state=123)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR=LinearRegression()
```

```python
LR.fit(X_train,Y_train)
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
pred=LR.predict(X_test)
```

```python
pred
```

# 7. Evaluate the Outcome

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
print("\t\t\t\tTesting Dataset")
print('--------------------------------------------------------------------')
print('Mean Absolute Error : ',mean_absolute_error(Y_test,pred))
print('Mean squared Error : ',mean_squared_error(Y_test,pred)**(1/2))
print('R squared score  : ',r2_score(Y_test,pred))
```

```python
print("\t\t\t\tTraining Dataset")
print('--------------------------------------------------------------------')
print('Mean Absolute Error : ',mean_absolute_error(Y_train,LR.predict(X_train)))
print('Mean squared Error : ',mean_squared_error(Y_train,LR.predict(X_train))**(1/2))
print('R squared score  : ',r2_score(Y_train,LR.predict(X_train)))
```
